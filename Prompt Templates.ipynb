```python
# Task 1: Replace Hardcoded Prompts
from langchain_core.prompts import PromptTemplate

template = "Explain {topic} for {audience} in a {tone} tone"

prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template
)

print(prompt.format(topic="AI", audience="beginners", tone="friendly"))
```

```python
# Task 2: Multi-Input Prompt System

template = "Explain {topic} for {audience} in a {tone} tone"

prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template
)

output1 = prompt.format(
    topic="AI",
    audience="beginners",
    tone="friendly"
)
print(output1)

output2 = prompt.format(
    topic="Python",
    audience="kids",
    tone="fun"
)
print(output2)

output3 = prompt.format(
    topic="Deep Learning",
    audience="engineers",
    tone="technical"
)
print(output3)
```

```python
# Task 3: Prompt Variations Engine

teaching_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

story_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print("Teaching Prompt:")
print(teaching_prompt.format(topic=topic))

print("\nInterview Prompt:")
print(interview_prompt.format(topic=topic))

print("\nStorytelling Prompt:")
print(story_prompt.format(topic=topic))
```

```python
# Task 4: ChatPrompt Template System

from langchain_core.prompts import ChatPromptTemplate

chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {role}."),
    ("user", "Explain {topic}")
])

role = "teacher"
topic = "Neural Networks"

messages = chat_template.format_messages(
    role=role,
    topic=topic
)

for msg in messages:
    print(msg.content)
```

```python
# Task 5: Input Validation Layer

def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience! Setting default = beginner")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone! Setting default = formal")
        tone = "formal"

    return audience, tone


aud, tone = validate_inputs("kids", "friendly")

print("Audience:", aud)
print("Tone:", tone)
```

```python
# Task 6: Prompt Generator App

def generate_prompt(topic, audience, tone, style):

    templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Ask questions about {topic} for {audience} in a {tone} interview style",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style"
    }

    if style not in templates:
        print("Invalid style! Using default = teaching")
        style = "teaching"

    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=templates[style]
    )

    return prompt.format(
        topic=topic,
        audience=audience,
        tone=tone
    )


result = generate_prompt(
    topic="Neural Networks",
    audience="beginners",
    tone="fun",
    style="storytelling"
)

print(result)
```

```python
# TASK 7: Template Reusablity Test

# ONE reusable template
template = "Explain {topic} for {audience} in a {tone} tone"

prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template
)

# 5 different inputs
inputs = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Data Science", "audience": "students", "tone": "casual"},
    {"topic": "Machine Learning", "audience": "engineers", "tone": "technical"},
    {"topic": "Cyber Security", "audience": "professionals", "tone": "formal"}
]

# Run loop
for i, inp in enumerate(inputs, 1):
    print(f"\nTest Case {i}:")
    print(prompt.format(**inp))
```
